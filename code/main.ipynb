## Importing the required modules

```python
import transport as tr
import matplotlib.pyplot as plt
import numpy as np
```

Initialize classes for calculating the characteristics of atomic and molecular nitrogen:

```python
N  = tr.Atom(14, 'N.txt')
N2 = tr.Molecule(28, 'N2.txt')
```

Constructing the procedure for drawing the necessary data:

```python
def PlotFigure(t, y1N, y2N, y1N2, y2N2, yLabel):
    plt.figure(figsize = (10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(t, y1N, color = 'blue', linestyle = 'solid', linewidth = 2, label = "Translational")
    plt.plot(t, y2N, color = 'red', linestyle = 'solid', linewidth = 2, label = "Internal")
    plt.grid(True)
    plt.xlabel('T, K')
    plt.ylabel(yLabel)
    plt.title('Atomic Nitrogen')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(t, y1N2, color = 'blue', linestyle = 'solid', linewidth = 2, label = "Translational")
    plt.plot(t, y2N2, color = 'red', linestyle = 'solid', linewidth = 2, label = "Internal")
    plt.grid(True)
    plt.xlabel('T, K')
    plt.ylabel(yLabel)
    plt.title('Molecular Nitrogen')
    plt.legend()    
    plt.show()
```

## An example of calculating specific heats

```python
t = np.linspace(500, 10000, 50)
shTrN = [N.SpecificHeatTr(x) for x in t]
shIntN = [N.SpecificHeatInt(x) for x in t]
shTrN2 = [N2.SpecificHeatTr(x) for x in t]
shIntN2 = [N2.SpecificHeatInt(x) for x in t]
```

```python
PlotFigure(t, shTrN, shIntN, shTrN2, shIntN2, 'Specific heat, J / (K * kg)')
```

## An example of calculating specific energies

```python
seTrN = [N.EnergyTr(x) for x in t]
seIntN = [N.EnergyInt(x) for x in t]
seTrN2 = [N2.EnergyTr(x) for x in t]
seIntN2 = [N2.EnergyInt(x) for x in t]
```

```python
PlotFigure(t, seTrN, seIntN, seTrN2, seIntN2, 'Specific energy, J / kg')
```

## An example of calculating statsums

```python
ssTrN = [N.StatSumTr(x) for x in t]
ssIntN = [N.StatSumInt(x) for x in t]
ssTrN2 = [N2.StatSumTr(x) for x in t]
ssIntN2 = [N2.StatSumInt(x) for x in t]
```

```python
PlotFigure(t, ssTrN, ssIntN, ssTrN2, ssIntN2, 'Statsum')
```
